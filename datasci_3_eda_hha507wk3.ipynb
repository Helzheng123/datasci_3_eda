```python
def clean_column_names(df):
    # Define a helper function to clean column names
    def clean_name(name):
        # Remove non-alphanumeric characters and convert to lowercase
        cleaned_name = re.sub(r'[^a-zA-Z0-9 -]', '', name)
        return cleaned_name.strip()  # Remove leading/trailing spaces and convert to lowercase

    # Rename columns using the helper function
    df.columns = [clean_name(col) if col.strip() in [
        'Life expectancy ',
        'Measles ',
        ' BMI ',
        'under-five deaths ',
        'Diphtheria ',
        ' HIV/AIDS',
        ' thinness 1-19 years',
        ' thinness 5-9 years'
    ] else col for col in df.columns]
    return df

# Apply the clean_column_names function to specific columns
data = clean_column_names(data)

data
```
