### Loading Packages and Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df
```

```python
# 20 Random data samples in the dataset
df.sample(20)
```

#### Cleaning Dataset

```python
# Here I will be checking the rows & columns

print("Number of rows: ", df.shape[0])
print("Number of columns: ", df.shape[1])
```

```python
# checking for any white space in the column names
df.columns
```

```python
# Renaming the column names that have extra white space

data = df.rename(columns = {'Life expectancy ' : 'Life expectancy', 'Measles ' : 'Measles', ' BMI ' : 'BMI', 'Diphteria ' : 'Diphteria', ' HIV/AIDS' : 'HIV/AIDS', ' thinness 1-19 years' : 'thinness 1-19 years', ' thinness 5-9 years' : 'thinness 5-9 years'})
data
```

```python
# Checking for the missing values in each
missing_values = data.isnull().sum()
missing_values
```

```python
# Replacing missing values with NaN
data.replace('', np.nan)
data
```

```python
# Remove any duplicates
data.drop_duplicates()
```

```python
# dropping any of the missing data
data.dropna()
```

```python
# Summary of Statistics using .describe()
data.describe()
```

```python
# summary statistics focusing on infant deaths & adult mortality
data[['infant deaths', 'Adult Mortality', 'Life expectancy']].describe()
```

#### Univariate Analysis

```python
# Calculating the measures of central tendency for infant deaths
# mean, median, mode

mean_infant_deaths = data['infant deaths'].mean()
median_infant_deaths = data['infant deaths'].median()
mode_infant_deaths = data['infant deaths'].mode().iloc[0]

print(f'Mean for infant deaths:', mean_infant_deaths)
print(f'Median for infant deaths:', median_infant_deaths)
print(f'Mode for infant deaths:', mode_infant_deaths)
```

```python
# Calculating the spread for infant deaths
# range, variance, standard deviation, IQR (interquartile range)

range_infant_deaths = data['infant deaths'].max() - data['infant deaths'].min()
variance_infant_deaths = data['infant deaths'].var()
std_dev_infant_deaths = data['infant deaths'].std()
iqr_infant_deaths = data['infant deaths'].quantile(0.75) - data['infant deaths'].quantile(0.25)

print(f'Range for infant deaths:', range_infant_deaths)
print(f'Variance for infant deaths:', variance_infant_deaths)
print(f'Standard Deviation for infant deaths:', std_dev_infant_deaths)
print(f'IQR for infant deaths:', iqr_infant_deaths)
```

```python
# Calculating the measures of central tendency for adult mortality
# mean, median, mode

mean_adult_mortality = data['Adult Mortality'].mean()
median_adult_mortality = data['Adult Mortality'].median()
mode_adult_mortality = data['Adult Mortality'].mode().iloc[0]

print(f'Mean for adult mortality:', mean_adult_mortality)
print(f'Median for adult mortality:', median_adult_mortality)
print(f'Mode for adult mortality:', mode_adult_mortality)
```

```python
# Calculating the spread for adult mortality
# range, variance, standard deviation, IQR (interquartile range)

range_adult_mortality = data['Adult Mortality'].max() - data['Adult Mortality'].min()
variance_adult_mortality = data['Adult Mortality'].var()
std_dev_adult_mortality = data['Adult Mortality'].std()
iqr_adult_mortality = data['Adult Mortality'].quantile(0.75) - data['Adult Mortality'].quantile(0.25)

print(f'Range for adult mortality:', range_adult_mortality)
print(f'Variance for adult mortality:', variance_adult_mortality)
print(f'Standard Deviation for adult mortality:', std_dev_adult_mortality)
print(f'IQR for adult mortality:', iqr_adult_mortality)
```

```python
# Calculating the measures of central tendency for life expectancy
# mean, median, mode

mean_life_expectancy = data['Life expectancy'].mean()
median_life_expectancy = data['Life expectancy'].median()
mode_life_expectancy = data['Life expectancy'].mode().iloc[0]

print(f'Mean for Life expectancy:', mean_life_expectancy)
print(f'Median for Life expectancy:', median_life_expectancy)
print(f'Mode for Life expectancy:', mode_life_expectancy)
```

```python
# Calculating the spread for life expectancy
# range, variance, standard deviation, IQR (interquartile range)

range_life_expectancy = data['Life expectancy'].max() - data['Life expectancy'].min()
variance_life_expectancy = data['Life expectancy'].var()
std_dev_life_expectancy = data['Life expectancy'].std()
iqr_life_expectancy = data['Life expectancy'].quantile(0.75) - data['Life expectancy'].quantile(0.25)

print(f'Range for Life expectancy:', range_life_expectancy)
print(f'Variance for Life expectancy:', variance_life_expectancy)
print(f'Standard Deviation for Life expectancy:', std_dev_life_expectancy)
print(f'IQR for Life expectancy:', iqr_life_expectancy)
```

### Visualize Distribution with Histograms

```python
# Histogram for Infant Deaths

plt.figure(figsize=(8,6))
plt.hist(data['infant deaths'], bins=30, density=True, alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Frequency Distribution for Infant Deaths')
plt.xlabel('Number of Infant Death per 1000 Population')
plt.ylabel('Frequency')
plt.grid(True)

x = np.linspace(min(data['infant deaths']), max(data['infant deaths']), 100)
pdf = (1/(std_dev_infant_deaths * np.sqrt(2*np.pi))) * np.exp(-(x - mean_infant_deaths)**2 / (2*std_dev_infant_deaths**2))
plt.plot(x, pdf, color='red', label='PDF')

plt.show()
```

```python
# Histogram for Adult Mortality

plt.figure(figsize=(8,6))
plt.hist(data['Adult Mortality'], bins=30, density=True, alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Frequency Distribution for Adult Mortality')
plt.xlabel('Number of Adult Mortality Rate (ages 15-60) per 1000 Population')
plt.ylabel('Frequency')
plt.grid(True)

x = np.linspace(min(data['Adult Mortality']), max(data['Adult Mortality']), 100)
pdf = (1/(std_dev_adult_mortality * np.sqrt(2*np.pi))) * np.exp(-(x - mean_adult_mortality)**2 / (2*std_dev_adult_mortality**2))
plt.plot(x, pdf, color='red', label='PDF')

plt.show()
```

```python
# Histogram for Life Expectancy

plt.figure(figsize=(8,6))
plt.hist(data['Life expectancy'], bins=30, density=True, alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Frequency Distribution for Life expectancy')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

x = np.linspace(min(data['Life expectancy']), max(data['Life expectancy']), 100)
pdf = (1/(std_dev_life_expectancy * np.sqrt(2*np.pi))) * np.exp(-(x - mean_life_expectancy)**2 / (2*std_dev_life_expectancy**2))
plt.plot(x, pdf, color='red', label='PDF')

plt.show()
```

```python
# Spearman correlation measuring correlation between Life Expectancy and Adult Mortality
spearman_corr = data[['Life expectancy','Adult Mortality']].corr(method='spearman')

print("Spearman Correlation Matrix:")
print(spearman_corr)

# Create a heatmap of the Spearman correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0)
plt.title("Spearman Correlation Heatmap")


plt.show()
```

### Observation:

Through the Spearman's correlation heatmap, there is a moderate negative correlation of -0.65 between Life Expectancy and Adult Mortality.

## Handling Outliers

```python
# Using IQR Method for adult mortality:

# Calculating quartiles and IQR

Variable1 = 'Adult Mortality'
q1 = data[Variable1].quantile(0.25)
q3 = data[Variable1].quantile(0.75)
iqr1 = q3 - q1

# Setting threshold for outlier detection

threshold_lower = q1 - 1.5 * iqr1
threshold_upper = q3 + 1.5 * iqr1

# Identifying outliers

outliers = [x for x in data['Adult Mortality'] if x < threshold_lower or x > threshold_upper]

# Printing outliers
print("Outliers:", outliers)
```

```python
# Using IQR Method for infant deaths:

# Calculating quartiles and IQR

Variable2 = 'infant deaths'
q1 = data[Variable2].quantile(0.25)
q3 = data[Variable2].quantile(0.75)
iqr2 = q3 - q1

# Setting threshold for outlier detection

threshold_lower = q1 - 1.5 * iqr2
threshold_upper = q3 + 1.5 * iqr2

# Identifying outliers

outliers = [x for x in data['infant deaths'] if x < threshold_lower or x > threshold_upper]

# Printing outliers
print("Outliers:", outliers)
```

```python
# Using IQR Method for life expectancy:

# Calculating quartiles and IQR

Variable3 = 'Life expectancy'
q1 = data[Variable3].quantile(0.25)
q3 = data[Variable3].quantile(0.75)
iqr3 = q3 - q1

# Setting threshold for outlier detection

threshold_lower = q1 - 1.5 * iqr3
threshold_upper = q3 + 1.5 * iqr3

# Identifying outliers

outliers = [x for x in data['Life expectancy'] if x < threshold_lower or x > threshold_upper]

# Printing outliers
print("Outliers:", outliers)
```
